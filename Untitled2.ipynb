```python
dataset_features = df.drop(['price','street','city','yr_built','statezip','country', 'date'], axis=1)
dataset_labels = df['price']
```

```python
dataset_features
```

```python
dataset_labels.head()
```

```python
from sklearn.model_selection import train_test_split
train_features,test_features,train_labels,test_labels=train_test_split(dataset_features,dataset_labels,test_size=0.2, random_state=200)
```

```python
#SOLVING FOR R SQUARED AND MSE
from sklearn import metrics
```

```python
def Predictive_Model(estimator):
    estimator.fit(train_features, train_labels)
    prediction = estimator.predict(test_features)
    print('R_squared:', metrics.r2_score(test_labels, prediction))
    print('Square Root of MSE:',np.sqrt(metrics.mean_squared_error(test_labels, prediction)))
    print('Mean Absolute Error:',metrics.mean_absolute_error( test_labels, prediction))
    print('Mean Squared Error:', metrics.mean_squared_error(test_labels,prediction))
    plt.figure(figsize=(10,5))
    sns.distplot(test_labels, hist=True, kde=False)
    sns.distplot(prediction, hist=True, kde=False)
    plt.legend(labels=['Actual Values of Price', 'Predicted Values of Price'])
    plt.xlim(0,)
def FeatureBar(model_Features, Title, yLabel):
    plt.figure(figsize=(10,5))
    plt.histplot(df.columns[df.columns!='Price'].values, model_Features)
    plt.xticks(rotation=45)
    plt.title(Title)
    plt.ylabel(yLabel)
```

```python
#USING RANDOM FOREST REGRESSOR
from sklearn.ensemble import RandomForestRegressor
rfg=RandomForestRegressor(n_estimators=200, random_state=0)
Predictive_Model(rfg)
```

```python
#USING LINEAR REGRESSION
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
Predictive_Model(lr)
```

```python
#USING RIDGE
from sklearn.linear_model import Ridge
rr = Ridge(alpha=100)
Predictive_Model(rr)
```

```python
#USING K NEIGHBOURS REGRESSOR
from sklearn.neighbors import KNeighborsRegressor
knn = KNeighborsRegressor(n_neighbors=5)
Predictive_Model(knn)
```

```python
#USING DECISION TREE REGRESSOR
from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor(max_depth=15, random_state=0)
Predictive_Model(dt)
```

```python
regressor = ['Random Forest Regressor','Linear Regression', 'Ridge Regression', 'KNN', 'Decision Tree']
models = [RandomForestRegressor(n_estimators=200, random_state=0) , LinearRegression(), Ridge(alpha=100), KNeighborsRegressor(n_neighbors=5), DecisionTreeRegressor(max_depth=15, random_state=0)]
R_squared = [] 
RMSE = []
MAE = []
MSE = [] 
for m in models:
    m.fit(train_features, train_labels)
    prediction_m = m.predict(test_features)
    r2 = metrics.r2_score(test_labels, prediction_m)
    rmse = np.sqrt(metrics.mean_squared_error(test_labels,prediction_m))
    mae = metrics.mean_absolute_error(test_labels, prediction_m)
    mse = metrics.mean_squared_error(test_labels,prediction_m)
    R_squared.append(r2)
    RMSE.append(rmse)
    MAE.append(mae)
    MSE.append(mse) 
basic_result = pd.DataFrame({'R squared':R_squared,'RMSE':RMSE,'MAE': MAE, 'MSE': MSE}, index=regressor)
basic_result
```

```python
basic_result.plot
```

```python
# Train classifiers
from sklearn.ensemble import VotingRegressor
reg1 = RandomForestRegressor(n_estimators=200, random_state=0)
reg2 = LinearRegression()
reg3 = Ridge(alpha=100)
reg4 = KNeighborsRegressor(n_neighbors=5)
reg5 = DecisionTreeRegressor(max_depth=15, random_state=0)

reg1.fit(train_features, train_labels)
reg2.fit(train_features, train_labels)
reg3.fit(train_features, train_labels)
reg4.fit(train_features, train_labels)
reg5.fit(train_features, train_labels)

ereg = VotingRegressor([("rfg", reg1), ("lr", reg2), ("rr", reg3), ("knn", reg4), ("dt", reg5) ])

ereg.fit(test_features, test_labels)
```

```python
xt = test_features[:20]

pred1 = reg1.predict(xt)
pred2 = reg2.predict(xt)
pred3 = reg3.predict(xt)
pred4 = reg4.predict(xt)
pred5 = reg5.predict(xt)
pred6 = ereg.predict(xt)
```

```python
plt.figure()
plt.plot(pred1, "gd", label="RandomForestRegressor")
plt.plot(pred2, "b^", label="LinearRegression")
plt.plot(pred3, "ys", label="Ridge")
plt.plot(pred4, "ro", label="KNeighborsRegressor")
plt.plot(pred5, "ko", label="DecisionTreeRegressor")
plt.plot(pred6, "r*", ms=10, label="VotingRegressor")

plt.tick_params(axis="x", which="both", bottom=False, top=False, labelbottom=False)
plt.ylabel("predicted")
plt.xlabel("training samples")
plt.legend(loc="best")
plt.title("Regressor predictions and their average")

plt.show()
```
